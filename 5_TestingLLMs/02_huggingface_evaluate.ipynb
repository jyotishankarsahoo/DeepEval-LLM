### Hugging Face Evaluate
- Accuracy
- Precision
- Recall
- F1 Score
- Perplexity

```python
import evaluate 
accuracy = evaluate.load("accuracy")
references = [1, 0, 1, 1]
predications = [0, 1, 1, 1]
results = accuracy.compute(references=references, 
                           predictions=predications)
print(f"Accuracy: {results["accuracy"]:.2%}")
```

```python
precision = evaluate.load("exact_match")
references = ["Execute", "Automation"]
predications = ["Execute", "Automations"]
results = precision.compute(references=references, 
                           predictions=predications)
print(f"Exact Match: {results["exact_match"]:.2%}")
```

```python
f1 = evaluate.load("f1")
# True label/Ground truth/reference
references = [1,0,1,1,0,1,0,1,1]
# Prediction from LLM
predications = [1,0,1,1,1,1,1,0,1]
results = f1.compute(references=references,
                     predictions=predications,
                     average="binary")
print(f"F1-Score: {results["f1"]:.2%}")
```
