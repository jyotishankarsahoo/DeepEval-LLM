#### Configure Local LLM using ChatOllama From langchain

```python
from langchain_ollama import ChatOllama
## Configure Local LLM using ChatOllama From langchain
llm = ChatOllama(
    base_url="http://localhost:11434",
    model="deepseek-r1:8b",
    temperature=0.5,
    num_predict=500 
)
```

#### Invoke the LLM

```python
llm.invoke("Who is the current president of united sates of america? Just give me the name and no explanation needed.").content
```
