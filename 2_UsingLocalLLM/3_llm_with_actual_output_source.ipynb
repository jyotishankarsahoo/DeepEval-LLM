#### Configure Local LLM using ChatOllama From langchain

```python
from langchain_ollama import ChatOllama
## Configure Local LLM using ChatOllama From langchain
llm = ChatOllama(
    base_url="http://localhost:11434",
    model="deepseek-r1:8b",
    temperature=0.5,
    num_predict=500
)
```

#### Testing with Actual Output Source

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.evaluate import evaluate, AsyncConfig
from deepeval.models import OllamaModel
import os

os.environ["DEEPEVAL_PER_ATTEMPT_TIMEOUT_SECONDS_OVERRIDE"] = "300"

local_model = OllamaModel(model="deepseek-r1:8b",
                          base_url="http://localhost:11434")

answer_relevancy_metric = AnswerRelevancyMetric(model=local_model)

test_case_1 = LLMTestCase(
    input="What is the capital of France?",
    actual_output=llm.invoke("What is the capital of France? Just give me the name and no explanation needed.").content
    )
test_case_2 = LLMTestCase(
    input="Who is the current president of united sates of america?",
    actual_output=llm.invoke("Who is the current president of united sates of america? Just give me the name and no explanation needed.").content
    )
evaluate(test_cases=[test_case_1, test_case_2], 
         metrics=[answer_relevancy_metric], 
         async_config=AsyncConfig(run_async=False))
```
