### Create confident AI Login

```python
import deepeval
import os
from dotenv import load_dotenv

load_dotenv(override=True)
API_KEY = os.getenv("CONFIDENT_API_KEY")
if not API_KEY:
    raise ValueError("API Key not found. Please set CONFIDENT_API_KEY in your .env file.")
deepeval.login(API_KEY)
```
