#### Creating Evaluation Dataset as Goldens in Confident AI

#### Creating Data set

```python
test_data = [
    {
        "input": "Who won 2025 Women's Cricket world cup",
        "expected_output": "India"
    },
    {
        "input": "Who introduced the GPT model",
        "expected_output": "OpenAI"
    },
    {
        "input": "What is the capital of France?",
        "expected_output": "Paris"
    }
]
```

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.evaluate import evaluate
from deepeval.dataset import EvaluationDataset, Golden

goldens = []
## Create Golden objects from test data
for data in test_data:
    golden = Golden(
        input=data["input"],
        expected_output=data["expected_output"],
    )
    goldens.append(golden)
## Create Evaluation Dataset
dataset = EvaluationDataset(goldens = goldens)
```

#### Push Data set to confident AI

```python
dataset.push(alias="test_qa_dataset")
```
