#### Answer Relevancy Metrics

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric

test_case = LLMTestCase(
    input="What is the capital of France?",
    actual_output="The capital of France is Paris.",
    retrieval_context=["Paris is the capital of France."]
)
answer_relevancy = AnswerRelevancyMetric()
answer_relevancy.measure(test_case)
print(f"Answer Relevancy Score: {answer_relevancy.score}")
```

#### Context Precision Metrics

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import ContextualPrecisionMetric

contextual_precision_metric = ContextualPrecisionMetric()
test_case = LLMTestCase(
    input="Who won 2025 Women World Cup?",
    # Should come from LLM or from agent or RAG system
    actual_output="India",
    # Should come from RAG - Vector DB, AI Agent - Agent's tool, LLM - LLM invoke response.
    retrieval_context=["India won the 2025 ICC women's world cup held in India"],
    expected_output="India has won the 2025 Women's World Cup.",
)
contextual_precision_metric.measure(test_case)
print("Contextual Precision Score:", contextual_precision_metric.score)
```
