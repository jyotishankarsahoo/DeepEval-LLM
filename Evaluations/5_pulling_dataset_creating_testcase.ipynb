### Preload Environment

```python
from dotenv import load_dotenv
load_dotenv(override=True)
```

#### Pull Dataset from Confident AI Dashboard

```python
from deepeval.dataset import EvaluationDataset
cloud_dataset = EvaluationDataset()
cloud_dataset.pull(alias="test_qa_dataset")
cloud_dataset
```

#### Prepare Testcases from Dataset

```python
from deepeval.test_case import LLMTestCase
## Dummy function to simulate LLM responses
def mock_llm_response(index):
    if index == 0:
        return "India"
    elif index == 1:
        return "OpenAI"
    else:
        return "Paris"

for index, golden in enumerate(cloud_dataset.goldens):
    test_case = LLMTestCase(
        input= golden.input,
        expected_output=golden.expected_output,
        actual_output=mock_llm_response(index)
    )
    # Append test cases to cloud dataset
    cloud_dataset.add_test_case(test_case)
```

#### Run Evaluations

```python
from deepeval.evaluate import evaluate
from deepeval.metrics import AnswerRelevancyMetric
## Run evaluation on the test cases
evaluate(cloud_dataset.test_cases, metrics=[AnswerRelevancyMetric()])
```
