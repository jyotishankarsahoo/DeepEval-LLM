### Answer relevancy using `Evaluate`

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.evaluate import evaluate

answer_relevancy_metric = AnswerRelevancyMetric()
test_case = LLMTestCase(
    input="Who own 2025 women's cricket world-cup",
    actual_output="India",
    retrieval_context=["India own women's world cup held in india"]
)
evaluate(test_cases=[test_case], metrics=[answer_relevancy_metric])
```

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.evaluate import evaluate

answer_relevancy_metric = AnswerRelevancyMetric()
test_case1 = LLMTestCase(
    input="Who own 2025 women's cricket world-cup",
    actual_output="India",
    retrieval_context=["India own women's world cup held in india"]
)
test_case2 = LLMTestCase(
    input="Who built the Claude model",
    actual_output="OpenAI",
    expected_output="Claude Anthropic",
    retrieval_context=["Claude Anthropic built the GPT models."]
)
evaluate(test_cases=[test_case1, test_case2], metrics=[answer_relevancy_metric])
```
