### Evaluate with Golden Dataset and Evaluation Dataset

```python
from deepeval.test_case import LLMTestCase
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.evaluate import evaluate
from deepeval.dataset import EvaluationDataset, Golden

# Create Golden instead of testcase
# This will never ever change
golden = Golden(
    input="Who own 2025 women's cricket world-cup",
    actual_output="India", # Comes From LLM or RAG or Chat Agent
    expected_output="India", # May Change
    context=["India own women's world cup held in india"]
)
dataset = EvaluationDataset()
dataset.add_golden(golden)
```

#### Creating Test case from Golden

```python
for golden in dataset.goldens:
    test_case = LLMTestCase(
        input=golden.input,
        actual_output=golden.actual_output,
        expected_output=golden.expected_output,
        retrieval_context=golden.context
    )
dataset.add_test_case(test_case)
```

```python
dataset.test_cases
```

#### Evaluate Test cases

```python
evaluate(test_cases=dataset.test_cases, metrics=[AnswerRelevancyMetric()])
```
